<h2 id="Dic">Diccionario</h2>

<h3 id="content">Que es un Diccionario?</h3>

Un diccionario consta de claves (keys) y valores (values). Es útil comparar un diccionario con una lista. En lugar de los índices numéricos, como una lista, los diccionarios tienen claves. Estas claves son las llaves que se utilizan para acceder a los valores dentro de un diccionario.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/DictsList.png" width="650" />

Un ejemplo de Diccionario <code>Dict</code>:

```python
# Creando un Diccionario

Dict = {"key1": 1, "key2": "2", "key3": [3, 3, 3], "key4": (4, 4, 4), ('key5'): 5, (0, 1): 6}
Dict
```

Las keys pueden ser "strings":

```python
# Accediendo a un valor con la clave "key1"

Dict["key1"]
```

Las claves también pueden ser cualquier objeto inmutable, como una tupla: 

```python
# Accediendo a un "valor" mediante la "key"

Dict[(0, 1)]
```

 Cada clave está separada de su valor por dos puntos  "<code>:</code>". Las comas separan los elementos y todo el diccionario está encerrado entre llaves. Un diccionario vacío sin elementos se escribe con solo dos llaves, como este  "<code>{}</code>".

```python
# Creando un directorio simple

release_year_dict = {"Thriller": "1982", "Back in Black": "1980", \
                    "The Dark Side of the Moon": "1973", "The Bodyguard": "1992", \
                    "Bat Out of Hell": "1977", "Their Greatest Hits (1971-1975)": "1976", \
                    "Saturday Night Fever": "1977", "Rumours": "1977"}
release_year_dict
```

En resumen, como una lista, un diccionario contiene una secuencia de elementos. Cada elemento está representado por una clave y su valor correspondiente. Los diccionarios se crean con dos llaves que contienen claves y valores separados por dos puntos. Para cada clave, solo puede haber un único valor; sin embargo, varias claves pueden contener el mismo valor. Las claves solo pueden ser cadenas, números o tuplas, pero los valores pueden ser de cualquier tipo de datos.

Es útil visualizar el diccionario como una tabla, como en la siguiente imagen. La primera columna representa las claves, la segunda columna representa los valores.

<h3 id="key">Keys</h3>

Puede recuperar los valores en función de los nombres:

```python
# Get value by keys

release_year_dict['Thriller'] 
```

This corresponds to: 


<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/DictsKeyOne.png" width="500" />

Igualmente para <b>The Bodyguard</b>

```python
# Get value by key

release_year_dict['The Bodyguard'] 
```

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/DictsKeyTwo.png" width="500" />

También te permite recuperar las claves del diccionario usando el método <code>release_year_dict()</code>:

```python
# Get all the keys in dictionary

release_year_dict.keys() 
```

Puedes recuperar los valores usando el método  <code>values()</code>:

```python
# Get all the values in dictionary

release_year_dict.values() 
```

Podemos añadir un elemento:

```python
# Agregar valor con clave en el diccionario

release_year_dict['Graduation'] = '2007'
release_year_dict
```

Podemos borrar elementos:   

```python
# Delete entries by key

del(release_year_dict['Thriller'])
del(release_year_dict['Graduation'])
release_year_dict
```

Podemos verificar si existe un elemento dentro del diccionario: 

```python
# Verify the key is in the dictionary

'The Bodyguard' in release_year_dict
```
