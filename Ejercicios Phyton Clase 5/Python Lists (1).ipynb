### Cosas que hacer con las listas

Hay muchas funciones integradas para trabajar con listas. Primero queremos determinar la longitud de una lista. En otros proyectos, conocer cuántos elementos tiene una lista. ¿Cuántos elementos tiene la lista ```bestestcolors```?

### Loops
A menudo queremos recorrer todos los elementos de una lista. Puede que recuerdes los ```Loops``` de la última vez. Ejecute el código en la celda de abajo.

````python
n=len(data)
for i in range(n):
  print ("Yo soy el elemento",i," y mi valor es de", data[i])
````

### Checking 

Imagina,que deseas recorrer todos los elementos de nuestra lista ```data``` como antes, pero solo imprimir el valor si es mayor que 10. Puede verificar usando el comando ```if``` como se muestra a continuación .

````python
for i in range(n):
  if(data[i]>10):
    print ("Found number greater than 10 with the value:",data[i])
````
